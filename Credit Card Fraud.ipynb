### Install Dependencies

### Balanced Dataset
    - 0 and 1 class - > 492 features each

```python
legit=legit.sample(n=fraud.shape[0])
```

```python
new_data=pd.concat([legit,fraud],axis=0)
```

```python
new_data.describe()
```

```python
new_data['Class'].value_counts()
```

### Model Select and Evaluation
    - Performed classifcation by using LR,KNN,DT,SVM

```python
X=new_data.drop(columns='Class',axis=1)
Y=new_data['Class']
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

### Scaling

```python
scaler = StandardScaler()
scaler.fit(X_train)
```

```python
X_train_scaled=scaler.transform(X_train)
X_test_scaled=scaler.transform(X_test)
```

```python
m1=LogisticRegression()
m2=svm.SVC(kernel='linear',C=10)
m3=DecisionTreeClassifier(max_depth=7)
m4=KNeighborsClassifier(n_neighbors=3)
```

```python
m1.fit(X_train_scaled,Y_train)
m2.fit(X_train_scaled,Y_train)
m3.fit(X_train_scaled,Y_train)
m4.fit(X_train_scaled,Y_train)
```

### Results
    - LR performed had decent training and testing accuracy
    - SVM and LR are similar
    - DT has overfitting issues, used h.p to reduced
    - KNN performs low on test dataset

```python
models=[m1,m2,m3,m4]
for model in models:
    X_train_pred=model.predict(X_train_scaled)
    training_accuracy=accuracy_score(X_train_pred,Y_train)
    X_test_pred=model.predict(X_test_scaled)
    test_accuracy=accuracy_score(X_test_pred,Y_test)
    print('{} \nTraining:{} Testing:{}\n'.format(model,training_accuracy,test_accuracy))


```
